### 做法其實很簡單，直接建立一個平均隨機取樣在[-2,2]的np.array。
### 全部算一次加起來然後除掉陣列長度在乘4回復原長

```python
X=np.random.uniform(-2,2,10000000)
intg = 0
for x in X: 
    intg += math.exp(x+x*x)
print(intg/(len(X))*4)
```

## Homework 5 problem12

### 好像就只要一直抽隨機np.array就好???

```python
def random_expeteion(nums):
    N = np.array([])
    for j in range(nums):
        sum = 0
        m = 1000000
        U = np.random.uniform(0,1,nums)
        for i in range(len(U)):
            sum += U[i]
            if sum > 1:
                m = min(m,i+1)
                break
        N = np.append(N,m)
    return np.mean(N)
```

```python
ans = random_expeteion(100)
print(f"ANS of 12.(a):{ans}")
```

```python
ans = random_expeteion(1000)
print(f"ANS of 12.(b):{ans}")
```
